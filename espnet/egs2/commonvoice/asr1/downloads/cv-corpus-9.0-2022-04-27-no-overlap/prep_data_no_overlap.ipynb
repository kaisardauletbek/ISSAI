```python
import re
def clean(string):
    string = string.lower()
    string = re.sub(r'[^ \w+]', '', string)
    return string

def clean_train(df_train, df_test, df_dev):
    train_path = list(df_train['path'])
    test_path = list(df_test['path'])
    dev_path = list(df_dev['path'])
    
    for j in range(len(train_path)):
        if train_path[j] in test_path or train_path[j] in dev_path:
            df_train.drop(index = j, inplace=True)
            
    df_train.reset_index(inplace=True)
    df_train.drop(columns = ['index'], inplace = True)


def rem_train_test_overlap(df_train, df_test):
    train = list(df_train['sentence'])
    test = list(df_test['sentence'])
    
    for i in range(len(test)):
        if test[i] in train:
            df_test.drop(index = i, inplace=True)
    df_test.reset_index(inplace=True)
    df_test.drop(columns = ['index'], inplace = True)
        
    
def rem_train_dev_overlap(df_train, df_dev):
    train = list(df_train['sentence'])
    dev = list(df_dev['sentence'])
    for i in range(len(dev)):
        if dev[i] in train:
            df_dev.drop(index = i, inplace=True)
    df_dev.reset_index(inplace=True)
    df_dev.drop(columns = ['index'], inplace = True)
```

```python
#azeri
az_validated = pd.read_csv(path + "az" + validated, sep = "\t")
az_dev = pd.read_csv(path + "az" + dev, sep = "\t")
az_test = pd.read_csv(path + "az" + test, sep = "\t")

#bashkir
ba_validated = pd.read_csv(path + "ba" + validated, sep = "\t")
ba_dev = pd.read_csv(path + "ba" + dev, sep = "\t")
ba_test = pd.read_csv(path + "ba" + test, sep = "\t")

#chuvash
cv_validated = pd.read_csv(path + "cv" + validated, sep = "\t")
cv_dev = pd.read_csv(path + "cv" + dev, sep = "\t")
cv_test = pd.read_csv(path + "cv" + test, sep = "\t")

#kazakh
kk_validated = pd.read_csv(path + "ksc1" + validated, sep = "\t")
kk_dev = pd.read_csv(path + "ksc1" + dev, sep = "\t")
kk_test = pd.read_csv(path + "ksc1" + test, sep = "\t")

#kyrgyz
ky_validated = pd.read_csv(path + "ky" + validated, sep = "\t")
ky_dev = pd.read_csv(path + "ky" + dev, sep = "\t")
ky_test = pd.read_csv(path + "ky" + test, sep = "\t")

#sakha
sah_validated = pd.read_csv(path + "sah" + validated, sep = "\t")
sah_dev = pd.read_csv(path + "sah" + dev, sep = "\t")
sah_test = pd.read_csv(path + "sah" + test, sep = "\t")

#tatar
tt_validated = pd.read_csv(path + "tt" + validated, sep = "\t")
tt_dev = pd.read_csv(path + "tt" + dev, sep = "\t")
tt_test = pd.read_csv(path + "tt" + test, sep = "\t")

#turkish
tr_validated = pd.read_csv(path + "tr" + validated, sep = "\t")
tr_dev = pd.read_csv(path + "tr" + dev, sep = "\t")
tr_test = pd.read_csv(path + "tr" + test, sep = "\t")

#uyghur
ug_validated = pd.read_csv(path + "ug" + validated, sep = "\t")
ug_dev = pd.read_csv(path + "ug" + dev, sep = "\t")
ug_test = pd.read_csv(path + "ug" + test, sep = "\t")

#uzbek
uz_validated = pd.read_csv(path + "uz" + validated, sep = "\t")
uz_dev = pd.read_csv(path + "uz" + dev, sep = "\t")
uz_test = pd.read_csv(path + "uz" + test, sep = "\t")
```

```python
languages = ["az", "ba", "cv", "ksc1", "ky", "sah", "tt", "tr", "ug", "uz"]
validated_lst = [az_validated, ba_validated, cv_validated, kk_validated, ky_validated, sah_validated, tt_validated, tr_validated, ug_validated, uz_validated]
dev_lst = [az_dev, ba_dev, cv_dev, kk_dev, ky_dev, sah_dev, tt_dev, tr_dev, ug_dev, uz_dev]
test_lst = [az_test, ba_test, cv_test, kk_test, ky_test, sah_test, tt_test, tr_test, ug_test, uz_test]
```

```python
for i in range(len(languages)):
    clean_train(validated_lst[i], test_lst[i], dev_lst[i])
#    rem_train_dev_overlap(validated_lst[i], dev_lst[i])
#    rem_train_test_overlap(validated_lst[i], test_lst[i])
```

```python
for i in range(len(languages)):
    c_dev = 0
    c_test = 0
    dev_id = []
    test_id = []
    dev_len_clean = []
    test_len_clean = []
    
    for j in range(len(dev_lst[i]['sentence'])):
        for sent_train in validated_lst[i]['sentence']:
            if sent_train == dev_lst[i]['sentence'][j]:
                c_dev+=1
                dev_id.append(j)
    dev_lst[i].drop(index = dev_id, inplace=True)
    dev_lst[i].reset_index(inplace=True)
    dev_lst[i].drop(columns = ['index'], inplace = True)
            
    for j in range(len(test_lst[i]['sentence'])):
        for sent_train in validated_lst[i]['sentence']:
            if sent_train == test_lst[i]['sentence'][j]:
                c_test+=1
                test_id.append(j)
    test_lst[i].drop(index = test_id, inplace=True)
    test_lst[i].reset_index(inplace=True)
    test_lst[i].drop(columns = ['index'], inplace = True)
            
    dev_pct = c_dev/len(validated_lst[i]) * 100
    test_pct = c_test/len(validated_lst[i]) * 100
    
    test_len_clean.append(len(test_lst[i]))
    dev_len_clean.append(len(dev_lst[i]))
    
    print(f"Language: {languages[i]}")
    print(f"Test: {len(test_lst[i])}\nDev:{len(dev_lst[i])}")
    print(f"Train / Test overlap: {test_pct}%")
    print(f"Train / Dev overlap: {dev_pct}%")
    print("\n\n")
```
